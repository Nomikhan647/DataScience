```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df1 = pd.read_csv("canada_per_capita_income.csv")
df1.head()
```

```python
df1.columns
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(df1.year, df1['per capita income (US$)'], color="blue")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df1[["year"]],df1['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Income")
plt.scatter(df1.year, df1['per capita income (US$)'], color="blue")
plt.plot(df1.year, reg.predict(df1[["year"]]), color="red")
```
