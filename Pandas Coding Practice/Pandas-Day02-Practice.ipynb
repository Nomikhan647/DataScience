Operations for DataFrame missing values

```python
#Operations for DataFrame Missing Values
dic = {'Name': ['Nauman', 'Asad', 'Luqman', 'Abbas'],
    'Age': [21,19,21,19],
    'Gender': ['Male','Male', 'Male', 'Male']}
index = [1,2,3,4]
df1 = pd.DataFrame(dic, index=index)
df1
```

```python
import numpy as np
dic2 = {'Name': ['Qayyum', 'Musa', 'Hamza', 'Hamid Raza', np.nan],
    'Age': [21,19,21,19,20],
    'Gender': ['Male','Male', 'Male', 'Male', 'Male']}
index = [1,2,3,4,5]
df2 = pd.DataFrame(dic2, index=index)
df2
```

```python
df3 = df2.copy() #creates a copy of DataFrame
df3.fillna('Mustafa') #fills null value with Mustafa
```

```python
df4 = df2.copy()
df4.dropna(how = 'any') #it will drop the row with missing values
```

```python
#DataFrame File Operations
df2.to_csv('names.csv') #saves the DataFrame as CSV file into file directory
```
